```python
spark.sql("CREATE SCHEMA IF NOT EXISTS workspace.ev_streaming")
spark.sql("CREATE VOLUME IF NOT EXISTS workspace.ev_streaming.raw_events")
print("Schema + volume ready ✅")
```

```python
import json, random, time, uuid
from datetime import datetime

VOLUME_PATH = "/Volumes/workspace/ev_streaming/raw_events"
stations = ["STN-01", "STN-02", "STN-03", "STN-04", "STN-05"]
cities   = ["Hyderabad", "Bengaluru", "Chennai", "Mumbai", "Delhi"]

def generate_batch(n=25):
    rows = []
    for _ in range(n):
        energy = round(random.uniform(2.0, 60.0), 2)
        rows.append({
            "session_id": str(uuid.uuid4()),
            "station_id": random.choice(stations),
            "city": random.choice(cities),
            "event_time": datetime.utcnow().isoformat(),
            "energy_kwh": energy,
            "cost_inr": round(energy * random.uniform(8, 12), 2),
            "duration_min": random.randint(10, 120),
            "status": random.choice(["completed", "completed", "completed", "interrupted"]),
        })
    return rows

batch = generate_batch(25)
fname = f"{VOLUME_PATH}/events_{int(time.time()*1000)}.json"
with open(fname, "w") as f:
    for row in batch:
        f.write(json.dumps(row) + "\n")

print(f"Wrote {len(batch)} events → {fname}")
```
